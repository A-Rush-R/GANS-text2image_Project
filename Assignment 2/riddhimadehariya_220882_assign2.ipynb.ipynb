```python
from sklearn.datasets import load_diabetes #We will use this dataset

#import libraries for numpy, Linear regression, r2_score, train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#split the dataset into Train(80%) and Test(20%)

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#fit X,y in Linear Regression Model of sklearn
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
X = X.reshape(-1, 1)
lin_reg.fit(X_train, Y_train)
```

```python
#predict on X_test and calculate r2_score

y_pred = lin_reg.predict(X_test)
from sklearn.metrics import r2_score
r2 = r2_score(Y_test, y_pred)
print(f"R-squared score: {r2}")
```

```python
#Write GDregressor class


class GDRegressor:
    
    def __init__(self,learning_rate,epochs):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X_train,y_train):
        # init your coefs
        num_samples, num_features = X_train.shape
        self.intercept_ = np.zeros(1) #initialise the intercept
        self.coef_ = np.zeros(num_features) #initialise the array of coefficients according to number of coefficients
        
        for i in range(self.epochs):
            y_pred = self.predict(X_train)
            # update all the coef and the intercept
            


            y_hat =self.coef_ *X + self.intercept_ #apply prediction formula i.e. (y_hat = coefficient*X + intercept) in terms of matrices coefficient and X   
            intercept_derivative =  np.mean(y_pred - y_train)#find derivative matrix 
            self.intercept_ = self.lr * intercept_derivative #update intercept using above intercept_derivative
            
            coef_derivative = np.dot(X_train.T, (y_pred - y_train)) / num_samples #derivative matrix for all coefficients in a single step
            self.coef_ = self.lr * coef_derivative#update intercept using coef_derivative
        
        #print the coefficients array and the intercept
    def print_coefficients(self):
        print("Coefficients:")
        print(self.weights)

    def print_intercept(self):
        print("Intercept:")
        print(self.bias)
    
    def predict(self,X_test):
        return np.dot(X_test, self.coef_) + self.intercept_ #new prediction in the form of (y = coefficients*X + intercept) using matrices
        
 
```

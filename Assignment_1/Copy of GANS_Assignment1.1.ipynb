```python
# Write a NumPy function to index and slice NumPy arrays similar to Python lists
# Show all three - indexing by a integer, indexing by a tuple, using range
import numpy as np
def indexing_example(arr):
    element = arr[2]
    #indexing by a integer
    print("Element at index 2:", element)

    #indexing by a tuple
    subarray = arr[1, 2]
    print("Element at index (1, 2):", subarray)

    #indexing by using a range
    sliced_array = arr[1:3, 0:2]
    print("Sliced array:\n", sliced_array)

a = np.array([[4,6,0], [0, 3, 1], [1, 8, 5]])
indexing_example(a)
```

```python
# Write a NumPy function to select values satisfying a specific condition like Values greater than 4 
import numpy as np
def value_greater_than(arr, given):
  mask = arr > given
  final_values = arr[mask]

  return final_values
a = np.array([[4,6,0], [0, 3, 1], [1, 8, 5]])
result = value_greater_than(a, 4)

print(result)

# Write a NumPy function to select values satisfying mulitple conditions like Values greater than 4 and divisible by 2
def value_with_conditions(arr):
  mask1 = arr > 4
  mask2 = arr % 2 == 0
  combined_mask = np.logical_and(mask1, mask2)
  
  final_array = arr[combined_mask]
  return final_array

a = np.array([[4,6,0], [0, 3, 1], [1, 8, 5]])

result = value_with_conditions(a)
print(result)

```

### Copying arrays
- new array object with a copy of the data
- Changes made in the copy array is not reflected in the original array

```python
a = np.array([[4,6,0], [0, 3, 1], [1, 8, 5]])
```

```python
# Write a NumPy function to copy the array 'a' and finally print an array where all elements are increased by 1
def copy_and_increase(arr):
  arr_copy = np.copy(arr)
  arr_copy += 1
  return arr_copy
   
result = copy_and_increase(a)
print(result)
```
